```python

names = []
emails = []


for i in all_profiles:
    print(i)

    driver.get(i + "overlay/contact-info/")
    time.sleep(4)
    soup = bs(driver.page_source)
    title_tag = soup.find('title')
    name = re.sub(r'^\(\d+\)\s+', '', title_tag.get_text().strip().split("|")[0].strip())
    print(name)
    names.append(name.split(" ")[0])
    email_link = soup.find('a', href=lambda href: href and href.startswith('mailto:'))
    # Check if the email link is found
    if email_link:
        email_address = email_link.get_text().strip()
        emails.append(email_address)
        print("Email Address:", email_address)
    else:
        print("Email address not found or profile does not have public contact info.")
        emails.append("0")


```

```python
tab = pd.DataFrame({"name": names, "email": emails})
```
